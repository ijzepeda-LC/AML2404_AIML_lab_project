```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Lista de películas y sus descripciones
peliculas = {
    "Titanic": "A love story on the tragic maiden voyage of the largest ship, the Titanic.",
    "Avatar": "A marine is sent to a distant planet inhabited by alien beings.",
    "The Godfather": "The story of a powerful mafia family and its leader, Don Vito Corleone.",
    "Maruchan": "La histoira de una sopa encantada",
    "Blanca nieves": "Un reino animado con 7 enanos",
    "pokemon": "Un mundo con animales que se capturan",
    "Digimon": "La historia de una poderosa familia con mascotas que se llaman digimones",
    "Mario": "La historia de un plomero que se pierde con su hermano y salva un reino fantastico",
    "Mario 4": "La historia de un plomero que vuelve al mundo para rescatar el mundo qeu ya habia rescatado"
    # Agrega más películas y descripciones aquí
}

# Crear listas de descripciones y etiquetas
descripciones = list(peliculas.values())
etiquetas = list(peliculas.keys())

# Tokenizar y secuenciar las descripciones
tokenizer = Tokenizer()
tokenizer.fit_on_texts(descripciones)
secuencias = tokenizer.texts_to_sequences(descripciones)
secuencias_pad = pad_sequences(secuencias)

# Crear el modelo de red neuronal
model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, 16, input_length=secuencias_pad.shape[1]),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(len(etiquetas), activation='softmax')
])

# Compilar el modelo
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entrenar el modelo
model.fit(secuencias_pad, tf.keras.utils.to_categorical(range(len(etiquetas)), num_classes=len(etiquetas)), epochs=10)

# Descripción para clasificar
descripcion = "Una película donde aparece un barco grande y navega por el hielo y hay una pareja de esposos."

# Preprocesar la descripción
secuencia_desc = tokenizer.texts_to_sequences([descripcion])
secuencia_desc_pad = pad_sequences(secuencia_desc, maxlen=secuencias_pad.shape[1])

# Clasificar la descripción
prediccion = model.predict(secuencia_desc_pad)
indice_prediccion = tf.argmax(prediccion, axis=1).numpy()[0]
mejor_pelicula = etiquetas[indice_prediccion]

# Imprimir el resultado
print("La mejor coincidencia es:", mejor_pelicula)
```
